```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df = pd.read_csv(r'./Hospital General Information.csv',sep=";", encoding='cp1252')
#df.drop(['Provider ID', ''], axis=1)
import csv
with open('Hospital General Information.csv', 'r', newline='') as file:
    csv_reader = csv.reader(file, delimiter=',', quotechar='"')
    df = pd.read_csv(file)   
```

```python
df = df[[
"Hospital Type","Hospital Ownership","Emergency Services","Hospital overall rating","Mortality national comparison",
"Safety of care national comparison","Readmission national comparison","Patient experience national comparison",
"Effectiveness of care national comparison","Timeliness of care national comparison","Efficient use of medical imaging national comparison"]]
df
```

```python
df.describe()
```

#Visualizing data

```python
for i, item in enumerate(df):
    plt.figure(figsize = (20, 6))
    sns.countplot(x = item, data = df)
    xt = plt.xticks(rotation=45)
    plt.xlabel(item)
    plt.ylabel('Count')
    plt.title(item)
```
